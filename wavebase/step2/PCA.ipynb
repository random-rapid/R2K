```python
import sklearn #機械学習のライブラリ
from sklearn.decomposition import PCA #主成分分析器
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine.txt", sep="\t", index_col=0)
df.head()
```

```python
from pandas import plotting 
plotting.scatter_matrix(df.iloc[:, 1:], figsize=(12, 12), c=list(df.iloc[:, 0]), alpha=0.5)
plt.show()
```

```python
# 行列の標準化
dfs = df.iloc[:, 1:].apply(lambda x: (x-x.mean())/x.std(), axis=0)
dfs.head()
```

```python
#主成分分析の実行
pca = PCA()
pca.fit(dfs)
# データを主成分空間に写像
feature = pca.transform(dfs)
```

```python
# 主成分得点
PCs = pd.DataFrame(feature, columns=[f"PC{x+1}" for x in range(len(dfs.columns))])
PCs.head()
```

```python
# 第一主成分と第二主成分でプロットする
plt.figure(figsize=(6, 6))
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, c=list(df.iloc[:, 0]))
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
plotting.scatter_matrix(pd.DataFrame(feature, 
                        columns=[f"PC{x+1}" for x in range(len(dfs.columns))]), 
                        figsize=(8, 8), c=list(df.iloc[:, 0]), alpha=0.5) 
plt.show()
```

```python
# 寄与率
evr = pd.DataFrame(pca.explained_variance_ratio_, index=[f"PC{x+1}" for x in range(len(dfs.columns))])
evr.head()
```

```python
# 累積寄与率を図示する
import matplotlib.ticker as ticker
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution rate")
plt.grid()
plt.show()
```

```python
# PCA の固有値
eigen = pd.DataFrame(pca.explained_variance_, index=[f"PC{x+1}" for x in range(len(dfs.columns))])
eigen.head()
```

```python
# PCA の固有ベクトル
eigen_vector = pd.DataFrame(pca.components_, columns=df.columns[1:], index=[f"PC{x+1}" for x in range(len(dfs.columns))])
eigen_vector.head()
```

```python
# 第一主成分と第二主成分における観測変数の寄与度をプロットする
plt.figure(figsize=(8, 8))
for x, y, name in zip(pca.components_[0], pca.components_[1], df.columns[1:]):
    plt.text(x, y, name)
plt.scatter(pca.components_[0], pca.components_[1], alpha=0.8)
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
