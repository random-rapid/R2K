```python
# ライブライのインポート
import power_spectrum as sp
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 画像ファイルの指定
img_file = r".\pictures\test4az.png"
```

```python
# 閾値の設定
thresh_val = 110
```

```python
# 2値化
img, retval = sp.thresh(img_file, thresh_val)
```

```python
# モノクロ画像とヒストグラムの表示
gray = sp.img2gray(img_file)

plt.figure(figsize=(18,6))
plt.subplot(1, 3, 1)
plt.hist(np.ravel(gray), 256)
plt.vlines(thresh_val, 0, max(np.bincount(np.ravel(gray))), "red", linestyles='dashed')

plt.subplot(1, 3, 2)
plt.imshow(gray, 'gray')

plt.subplot(1, 3, 3)
plt.imshow(img, 'gray')
plt.show()
plt.close()
```
